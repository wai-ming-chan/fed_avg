```python
# import libraries
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets
from torchvision.transforms import ToTensor, Lambda
import matplotlib.pyplot as plt
import numpy as np

#-----------------------------------------------------------------------

#--------------------------------------------

import pickle # library for save/load variable
```

```python
f = open('figs/MNIST_2NN_Centralized.pkl', 'rb')
unpickled = np.load(f, allow_pickle = True)
f.close()

array = unpickled
result_cen_lossTT_vals = array[0][:]
result_cen_lossTR_vals = array[1][:]
result_cen_accu_vals = array[2][:]


f = open('figs/MNIST_2NN_FedAvg.pkl', 'rb')
unpickled = np.load(f, allow_pickle = True)
f.close()

array = unpickled
result_fedavg_lossTT_vals = array[0][:]
result_fedavg_lossTR_vals = array[1][:]
result_fedavg_accu_vals = array[2][:]

f = open('figs/MNIST_2NN_FedSGD.pkl', 'rb')
unpickled = np.load(f, allow_pickle = True)
f.close()

array = unpickled
result_fedsgd_lossTT_vals = array[0][:]
result_fedsgd_lossTR_vals = array[1][:]
result_fedsgd_accu_vals = array[2][:]

```

```python
plt.figure(figsize=(5,5))
plt.title("Training Loss")
plt.plot(result_fedavg_lossTR_vals,label="FedAvg [B=10, C=0.1]")
plt.plot(result_fedsgd_lossTR_vals,label="FedSGD [B=infy, C=0]")
plt.plot(result_cen_lossTR_vals,label="CenSGD [B=64]")
# plt.plot(losses_test,label="test")
plt.xlabel("Rounds")
plt.ylabel("Loss")
plt.legend()
plt.savefig('figs/All_Loss.png')
plt.show()

plt.figure(figsize=(5,5))
plt.title("Test Data Accuracy")
# plt.plot(acc_train,label="train")
plt.plot(result_fedavg_accu_vals,label="FedAvg [B=10, C=0.1]")
plt.plot(result_fedsgd_accu_vals,label="FedSGD [B=infy, C=0]")
plt.plot(result_cen_accu_vals,label="CenSGD [B=64]")
plt.xlabel("Rounds")
plt.ylabel("Accuracy")
plt.legend()
plt.savefig('figs/All_Accuracy.png')
plt.show()
```
